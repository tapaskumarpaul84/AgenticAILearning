```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings=OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
text="I live in Kolkata the city of joy."

embed_text=embeddings.embed_query(text)
```

```python
embed_text
```

```python
len(embed_text)
```

```python
from langchain_community.document_loaders import TextLoader

txt_loader=TextLoader("speech.txt")

txt_docs=txt_loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

txt_spliter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)

split_docs=txt_spliter.split_documents(txt_docs)

split_docs
```

```python
split_docs[0].page_content
```

```python
embed_text=embeddings.embed_query(split_docs[0].page_content)
embed_text
```

```python
len(embed_text)
```
